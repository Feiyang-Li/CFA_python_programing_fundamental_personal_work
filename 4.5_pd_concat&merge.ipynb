Now that you learned about concatenation and merging of pandas DataFrames, try the following practice opportunity.

Assume that you became a new client of an investment firm that has its data stored in “investor_df.” Write a Python code that performs the following:

Define a new pandas DataFrame named “new_client_df” that contains your information, such as client ID (choose C1), first name, last name, age, portfolio size, years with investment firm, risk tolerance, goal, salaries, and job title.

Concatenate this new DataFrame to the original pandas DataFrame “investor_df.”

Reset the index.

```python
import pandas as pd

investor_df = pd.read_csv("investors_data.csv")
```

```python
myInfo = pd.DataFrame({
    "First Name": ["Fion"],
    "Last Name": ["Li"],
    "client ID": ["12345"],
    "Age": [128],
    "Portfolio Size": [1000],
    "Years with Investment Firm": [0],
    "Risk Tolerance": ["conservative"],
    "salaries": [0],
    "Job Title": ["Finance analytic"]
})
```

```python
myInfo
```

```python
new_df = pd.concat([investor_df, myInfo], ignore_index=True)
```

```python
new_df
```

```python
# we only care about investor_df, not necessary myInfo
myInfo_filtered = myInfo[investor_df.columns.intersection(myInfo.columns)]
myInfo_filtered
```

```python
better_new_df = pd.concat([investor_df, myInfo_filtered], ignore_index=True)
better_new_df
```
