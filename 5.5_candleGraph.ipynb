```python
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime
import pandas as pd
import numpy as np
```

```python
JPM_df = pd.read_csv("JPM.csv")
JPM_df.iloc[:, 0:6]
```

```python
NewJPM_df = JPM_df.set_index("Date")
```

```python
import cufflinks as cf
# a simplification of pyplot which is a addon to matplot

cf.go_offline()
```

```python
fig = cf.QuantFig(NewJPM_df.iloc[:, 0:5], title="Candlestick for the JPM stock price", name="JPM")
fig.add_sma(periods=[30, 100], column="Close", color=['magenta', 'green'])
fig.iplot(theme = 'white', up_color = 'green', down_color = 'red')
```

```python
# chat gpt generated, I am trying to figure out how it work. 

import plotly.graph_objects as go

fig = go.Figure(data=[
    go.Candlestick(
        x=NewJPM_df.index,
        open=NewJPM_df["Open"],
        high=NewJPM_df["High"],
        low=NewJPM_df["Low"],
        close=NewJPM_df["Close"],
        increasing_line_color='green',
        decreasing_line_color='red',
        name="JPM"
    )
])

fig.update_layout(
    title="Candlestick for the JPM stock price",
    xaxis_title="Date",
    yaxis_title="Price"
)

NewJPM_df["SMA_30"] = NewJPM_df["Close"].rolling(30).mean()
NewJPM_df["SMA_100"] = NewJPM_df["Close"].rolling(100).mean()

fig.add_trace(go.Scatter(
    x=NewJPM_df.index,
    y=NewJPM_df["SMA_30"],
    mode="lines",
    name="SMA 30"
))

fig.add_trace(go.Scatter(
    x=NewJPM_df.index,
    y=NewJPM_df["SMA_100"],
    mode="lines",
    name="SMA 100"
))


fig.show()
```
