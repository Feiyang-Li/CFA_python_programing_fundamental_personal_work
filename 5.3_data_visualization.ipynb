```python
def get5yrStockPriceHistory(stockName):  # our old function
    now = datetime.now()
    fiveYrBefore = now - relativedelta(years=5)
    stock = yf.Ticker(stockName)
    return stock.history(start=fiveYrBefore, end=now)  
```

```python
def plt_finance_data(df, x_axis, include_rows, **kwargs):
    ## this df is always from yfinance
    df_n = df.reset_index()
    print(df_n)
    fig = px.line()
    for i in df_n.columns[include_rows]:
        fig.add_scatter(x = df_n[x_axis], y = df_n[i], name=i)
    fig.update_traces(line_width = 5)
    fig.update_layout(plot_bgcolor="white", **kwargs)
    fig.show()
```

```python
plt_finance_data(JPM_df_5, "Date", include_rows= slice(1, 5), title= "JPM open/close/average diplay")
```
