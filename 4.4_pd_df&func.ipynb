Now that you have learned how to use functions with a pandas DataFrame, try the following practice opportunity.

Write some Python code that performs the following tasks:

Import the “investors_data.csv” dataset and place it in “investor_df” DataFrame using pandas.

Define a function named “ranking” that receives the portfolio size and adds one of the following ranks:

Low: Client portfolio size is less than $100,000

Medium: Client portfolio size is between $100,000 and $500,000

High: Client portfolio size is more than $500,000

Apply the function to the “investor_df” DataFrame, and add the output to a new column titled “Rank.”

```python
import pandas as pd
investor_df = pd.read_csv("investors_data.csv")

def ranking(k):
    # here k is a series, a row of the df
    item = k["Portfolio Size"]
    if item < 100000:
        return "Low"
    elif item > 500000:
        return "High"
    return "Medium"

investor_df["Rank"] = investor_df.apply(ranking, axis=1)
```
